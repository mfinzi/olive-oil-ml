## BGAN

```python
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
```

```python
trainset[0][0].size()
```

```python
def weight_init(m):
    if isinstance(m, nn.ConvTranspose2d) or isinstance(m, nn.Conv2d):
        m.weight.data.normal_(0.0, 0.02)
        m.bias.data.zero_()
        
def gen_weight_init(m):
    if isinstance(m, nn.ConvTranspose2d) or isinstance(m, nn.Conv2d):
        m.weight.data.normal_(0.0, 0.02)
        m.bias.data.zero_()
        
        
class Expression(nn.Module):
    def __init__(self, func):
        super(Expression, self).__init__()
        self.func = func
        
    def forward(self, input):
        return self.func(input)        
        
class GeneratorNetwork(nn.Module):
    # initializers
    def __init__(self, d=128, z_dim=100):
        super(GeneratorNetwork, self).__init__()
        self.input_dim = z_dim
        self.core_net = nn.Sequential(
            nn.Linear(z_dim, 4*4*(4*d)), nn.BatchNorm1d(4*4*(4*d)), nn.ReLU(),
            Expression(lambda tensor: tensor.view(tensor.size(0),4*d,4,4)),
            nn.ConvTranspose2d(4*d,2*d,5,2,2,1), nn.BatchNorm2d(2*d), nn.ReLU(),
            nn.ConvTranspose2d(2*d,  d,5,2,2,1), nn.BatchNorm2d(d),   nn.ReLU(),
            nn.ConvTranspose2d(d  ,  3,5,2,2,1), nn.Tanh(),
        )

    # forward method
    def forward(self, z):
        return self.core_net(z)
    

class DiscriminatorNetwork(nn.Module):
    
    def __init__(self, d=64,numClasses=1):
        super(DiscriminatorNetwork, self).__init__()
        self.numClasses = numClasses
        
        self.core_net = nn.Sequential(
            nn.Conv2d(  3,  d,3,1,1), nn.BatchNorm2d(  d), nn.LeakyReLU(0.2),
            nn.Conv2d(  d,  d,3,1,1), nn.BatchNorm2d(  d), nn.LeakyReLU(0.2),
            nn.Conv2d(  d,  d,3,2,1), nn.BatchNorm2d(  d), nn.LeakyReLU(0.2),
            #nn.Dropout2d(0.5),
            nn.Conv2d(  d,2*d,3,1,1), nn.BatchNorm2d(2*d), nn.LeakyReLU(0.2),
            nn.Conv2d(2*d,2*d,3,1,1), nn.BatchNorm2d(2*d), nn.LeakyReLU(0.2),
            nn.Conv2d(2*d,2*d,3,2,1), nn.BatchNorm2d(2*d), nn.LeakyReLU(0.2),
            #nn.Dropout2d(0.5),
            nn.Conv2d(2*d,2*d,3,1,0), nn.BatchNorm2d(2*d), nn.LeakyReLU(0.2),
            nn.Conv2d(2*d,2*d,1,1,0), nn.BatchNorm2d(2*d), nn.LeakyReLU(0.2),
            nn.Conv2d(2*d,2*d,1,1,0), nn.BatchNorm2d(2*d), nn.LeakyReLU(0.2),
            Expression(lambda tensor: tensor.mean(3).mean(2).squeeze()),
            nn.Linear(2*d,numClasses),
        )

    # forward method
    def forward(self, x):
        return self.core_net(x)
```

```python
gen = GeneratorNetwork()
disc = DiscriminatorNetwork()

z = Variable(torch.from_numpy(np.random.rand(10, 100)).float())
print(gen.forward(z).size())
print(disc(gen.forward(z)).size())
```

```python
from bgan.bgan import BGAN
```

```python
batch_size = 50
z_dim = 100
```

```python
#prior = UninformativePrior()
prior = FactorizedNormalPrior(std=10.)
gen = GeneratorNetwork()
disc = DiscriminatorNetwork()
# gen.apply(weight_init)
gen.apply(gen_weight_init)
disc.apply(weight_init)
bgan = BGAN(gen, disc, prior, alpha=.9, eta=0.00001, disc_lr = 0.00002, gen_observed=20000, MAP=False, cuda=True)
samples = []
```

```python
for param in gen.parameters():
    print(param.size(), np.linalg.norm(param.cpu().data.numpy()))
```

## Training

```python
num_samples = 10
```

```python
if isinstance(samples, np.ndarray):
    samples = [samples]
    
for epoch in range(100):
    for i, data_ in enumerate(trainloader, 0):
        batch = data_[0].float()
        bgan.step(batch)
    if not epoch%1:
        batchv = Variable(batch)
        d_loss, g_loss = bgan.loss(batchv.cuda())
        print('\t', d_loss.cpu().data.numpy()[0], g_loss.cpu().data.numpy()[0])
    if not epoch%1:
        print(epoch)
    if not epoch%1:
        samples.append(bgan.sample(num_samples).cpu().data.numpy())
samples = np.vstack(samples)
```
